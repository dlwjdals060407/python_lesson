```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("data.csv", encoding='cp949')
df
```

```python
categoryDf = [df['구분'] == '발생건수']
```

```python
categoryDf = categoryDf.sort_values(by='연도')
x = categoryDf['연도']
y = categoryDf['해킹']
plt.plot(x, y)
plt.show()
```
